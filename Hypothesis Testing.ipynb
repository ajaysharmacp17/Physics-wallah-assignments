## Q.1. What is a null hypothesis (H₀) and why is it important in hypothesis testing?
Ans. The Null Hypothesis (H₀)
The Null Hypothesis (H₀) is a formal statement asserting that there is no significant difference, effect, or relationship between specified populations or variables. It represents the status quo or a position of no change.
Importance: It provides a concrete baseline for statistical testing. In the scientific method, we do not aim to prove the alternative hypothesis directly; instead, we attempt to find sufficient evidence to reject the null hypothesis. If the evidence is insufficient, we "fail to reject" H₀.

## Q.2. What does the significance level (α) represent in hypothesis testing?
Ans. The Significance Level (α) represents the probability of committing a Type I error— rejecting the null hypothesis when it is actually true.
- It is the threshold for deciding whether a result is "statistically significant."
- A common value is 0.05 (5%), meaning there is a 5% risk of concluding a difference exists when it does not.

## Q.3. Differentiate between Type I and Type II errors.
Ans. Type I Error (False Positive): Occurs when H₀ is rejected despite being true. You conclude an effect exists when it doesn't.
     Type II Error (False Negative): Occurs when you fail to reject H₀ despite it being false. You miss a real effect or difference.

## Q.4. Explain the difference between a one-tailed and two-tailed test. Give an example of each.
Ans. One-Tailed Test: Used when the hypothesis predicts a specific direction of the effect (e.g., "Productivity will increase"). The critical region is in one tail of the distribution.
Two-Tailed Test: Used when the hypothesis predicts a change or difference but doesn't specify the direction (e.g., "There is a difference in mean time"). The critical region is split between both tails.

## Q.5. A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.
Ans. Claim: Mean time (μ) = 10 minutes. Sample (n=9) mean = 12, Std Dev (s) = 3, α = 0.05.
H₀: μ = 10
Hₐ: μ ≠ 10 (Two-tailed)
Test Selection: Since n < 30 and population σ is unknown, a t-test is required.
Degrees of Freedom: df = 9 - 1 = 8.
Analysis: The critical t-value for 8 df at 0.05 is ±2.306. The calculated t-value is (12-10)/(3/√9) = 2.0. Since 2.0 < 2.306, we fail to reject the null hypothesis.

Ans. Condition                                             Test
