# Probability Basics
## Question 1: A die is rolled. What is the probability of getting:
Ans. (a) An even number
Even numbers on a die are: {2,4,6}
Probability formula:
P(Event) = (Number of favorable outcomes) / (Total outcomes)
Number of favorable outcomes = 3
3/6 = 1/2

(b) A number greater than 4
Numbers greater than 4 are: {5,6}
Number of favorable outcomes = 2
P(number > 4) = 2/6 = 1/3

## Question 2: In a class of 50 students:
20 like Mathematics (M)
15 like Science (S)
5 like both subjects
What is the probability that a student chosen at random likes Mathematics or Science?

Ans. Total students = 50
Students who like Mathematics (M) = 20
Students who like Science (S) = 15
Students who like both (M ∩ S) = 5
P(M ∪ S) = P(M) + P(S) - P(M ∩ S)
probability = (M + S - both) / total_students
Probability = 30/50 = 0.6 (or 3/5)

## Question 3: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability that the next ball is also red (without replacement)?
Ans. Total balls initially = 5 (3 Red, 2 Blue)
After drawing 1 Red ball, remaining Red balls = 2
Remaining Total balls = 4
Probability (2nd is Red | 1st was Red) = 2/4 = 0.5

## Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?
Ans. Stratified Sampling
Stratified sampling ensures that subgroups (strata) of a population are represented in the sample. By dividing the population into "Boys" and "Girls" strata, you can specifically select an equal number from each to guarantee equal representation, whereas Simple Random Sampling might result in an accidental imbalance.

## Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height = 158 cm.
Find the sampling error.
Ans. Population mean height (N=1000) = 160 cm. Sample mean height (n=100) = 158 cm.
Sampling Error = Population Mean - Sample Mean
Sampling Error = 160 cm - 158 cm = 2cm.

## Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what is the standard error of the mean (SEM)?
Ans. Population Standard Deviation: ₹5,000 , Sample Size (n): 100 employees
Square root of the sample size: 100 = 10
population standard deviation by that result: SEM = 5000/10
SEM = 500

## Question 7: In a group of 100 students:
40 like Cricket (C)
30 like Football (F)
10 like both Cricket and Football
Find the probability that a student likes at least one sport.
Ans. total = 100
P(C ∪ F) = P(C) + P(F) - P(C ∩ F)
P(C ∪ F) = 40/100 + 30/100 - 10/100 = 0.6

## Question 8: From a deck of 52 cards, two cards are drawn without replacement. What is the probability that both are Aces?
Ans. P(1st Ace) = 4/52
P(2nd Ace | 1st was Ace) = 3/51
P(Both Aces) = 4/52 * 3/51 = 12/2652 = 0.0045

## Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?
Ans. P(Defective) = 0.02, so P(Non-defective) = 1 - 0.02 = 0.98
For 5 bulbs to be non-defective: 0.98^5 = 0.9039 or 90.39 %
