## Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.
Ans. Covariance: Measures the direction of the linear relationship between two variables. It indicates whether variables tend to increase or decrease together.
Correlation: A standardized version of covariance that measures both the direction and the strength of the linear relationship.

Differences: Scale: Covariance can range from -∞ to +∞, making it sensitive to the scale of the variables. Correlation is scaled between -1 and +1.

Interpretation: Because it is dimensionless, correlation is easier to interpret across different datasets than covariance.

## Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?
Ans. Positive Covariance: Indicates that as one variable increases, the other tends to increase as well.
Negative Covariance: Indicates that as one variable increases, the other tends to decrease.
Zero Covariance: Indicates that there is no linear relationship between the two variables.

## Q3.Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?
Ans. Limitations of Covariance: Its value depends on the units of measurement (e.g., changing from meters to centimeters will change the covariance value), making it difficult to compare relationships.

Why Correlation is Preferred: It provides a universal scale (-1 to +1), allowing researchers to compare the strength of relationships between different pairs of variables regardless of their units.

## Q4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?
Ans. Pearson Correlation: Measures the linear relationship between continuous variables.

Spearman Rank Correlation: Measures the monotonic relationship (whether variables move in the same direction, but not necessarily at a constant rate) using the ranks of the data.

Preference: Use Spearman when the data is ordinal (ranked), non-normally distributed, or contains outliers that would skew a linear analysis.

## Q5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?
Ans. Interpretation: A coefficient of $0.85$ indicates a strong, positive linear relationship between X and Y.

Causation: No, you cannot infer causation. Correlation only measures association; it does not prove that changes in X cause changes in Y, as a third "lurking" variable could be influencing both.
